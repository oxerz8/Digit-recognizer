```python
import numpy as np
import pandas as pd
import sklearn.neural_network as nn
```

```python
train_data = pd.read_csv('train.csv')
Xtest = pd.read_csv('test.csv')
```

```python
train_data.head()
```

```python
Xtrain = train_data.drop(['label'],axis='columns')
Ytrain = train_data['label']
```

```python
Xtrain.head()
```

```python
Ytrain.head()
```

```python
clf = nn.MLPClassifier(hidden_layer_sizes=5, activation='logistic', max_iter=1000, solver='sgd', tol=10**(-6), learning_rate_init=0.01)
```

```python
clf.fit(Xtrain,Ytrain)
```

```python
clf.score(Xtrain,Ytrain)
```

```python
try1 = clf.predict(Xtest)
```
